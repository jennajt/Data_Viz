Plotting the following functions:

$$y1 = x^2$$

$$y2 = \log x$$

$$y3 = e^x$$

$$y4 = \frac{e^x}{e^x + 1}$$


```python
import math
x_data = np.arange(-10,11,2)

def y1(x_data):
    return x_data**2

def y2(x_data):
    if x_data<=0:
        return 0
    else:
        return math.log(x_data)

def y3(x_data):
    return math.exp(x_data)

def y4(x_data):
    return math.exp(x_data)/((math.exp(x_data))+1)


y1_values = list(map(y1,x_data))
y2_values = list(map(y2,x_data))
y3_values = list(map(y3,x_data))
y4_values = list(map(y4,x_data))

plt.figure(figsize=(10,5))

plt.plot(x_data,y1_values)
plt.plot(x_data,y2_values)
plt.plot(x_data,y3_values)
plt.plot(x_data,y4_values)

plt.ylim(-2,6)
plt.xlim(-4,4)

plt.show()
```

## Formatting basics

```python
from sklearn import datasets

x, y, _ = datasets.make_regression(n_samples=100, n_features=1, noise=10, coef=True, random_state=0)

plt.scatter(x,y);
plt.plot([-3, 2.5],[-150,100]);
```

```python
from sklearn import datasets

x, y, _ = datasets.make_regression(n_samples=100, n_features=1, noise=10, coef=True, random_state=0)

plt.scatter(x,y,label="Scatter");
plt.plot([-3, 2.5],[-150,100],label="Regression Line");
plt.title('Linear Regression')
plt.legend()
plt.xlabel('input value')
plt.ylabel('output value')

plt.annotate("Regression line", (0,0), (1,-50), arrowprops={"arrowstyle":"->"})
```

## Subplotting

```python
bitcoin = [3,3,4,5,5,6,11,12,13,14,16,14,18,15,19,18,19,15,10,8,9,9,9,12,13]
ethereum = [1,1,1,2,2,3,3,4,4,4,4,8,9,12,10,9,14,15,10,8,6,5,5,8,7]
euro = [7,7,7,7.6,7,7.8,7,7,7,7,7.6,7,7,7,7,6.8,6.9,7,7,7,7,7,7.2,7.2,7.3]


wallet_2015 = {"currency": ["bitcoin","ethereum","euro"], "value_in_euro": [234,74,530]}
wallet_2020 = {"currency": ["bitcoin","ethereum","euro"], "value_in_euro": [234*4,74*7,530*1.3]}

# Create figure
fig = plt.figure()

#first sub-plot
ax1 = fig.add_subplot(1,2,1)
ax1.bar(wallet_2015["currency"], wallet_2015["value_in_euro"])

#second sub-plot
ax2 = fig.add_subplot(1,2,2)
ax2.bar(wallet_2020["currency"], wallet_2020["value_in_euro"])
plt.show()

#third line plot
plt.plot(bitcoin)
plt.plot(ethereum)
plt.plot(euro)
plt.show();
```

```python
# Create Complex Layout

fig = plt.figure(figsize = (15,10))
plt.title('Empty Grid Spec')
gs= fig.add_gridspec(3,3)
ax1 = fig.add_subplot(gs[:2,:2])
ax2 = fig.add_subplot(gs[-1,-1])
ax3 = fig.add_subplot(gs[1,-1])
ax4 = fig.add_subplot(gs[2,0])
ax5 = fig.add_subplot(gs[2,1])
ax6 = fig.add_subplot(gs[2,-1])
ax7 = fig.add_subplot(gs[0,2])

plt.show();
```
