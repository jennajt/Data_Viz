# Seaborn Basics - Exploring the Invoices and Tips Dataset

```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

## Load dataset

```python
df = sns.load_dataset('tips')

df.head()
```

## Exploring the data

```python
sns.countplot(data=df,x='day',palette="Set3").set(title='Number of reservations');
```

**Saturday is the most popular day for reservations**

```python
sns.countplot(data=df,x='day',hue='smoker',palette="Set3").set(title='Number of reservations - smokers/non-smokers');
```

**There are more smokers on Saturday**

```python
sns.histplot(df['tip'],kde=True).set(title='Distribution of tips');
```

**Most people leave ~$2 in tips**

```python
sns.catplot(x = 'day', y='tip', kind='box', data=df).set(title='Distribution of tips');
```

**Greatest distribution in tips on Sunday, many outliers on Saturday**

```python
g = sns.FacetGrid(df, col="day", row="smoker", hue="smoker")
g.map(sns.scatterplot,'tip','total_bill');
```

```python
sns.regplot(x='total_bill', y='tip', data=df).set(title='Regression of tips on total bill');
```

```python
sns.jointplot(x='total_bill', y='tip', kind='scatter',data=df);
```

```python
g = sns.FacetGrid(df, col='smoker')
g.map(sns.regplot, 'total_bill','tip');
```

**Non-smokers leave bigger tips**

```python
sns.pairplot(df);
```
