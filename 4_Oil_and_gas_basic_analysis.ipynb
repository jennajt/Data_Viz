# U.S. Oil and Gas Production - Basic Analysis

Analysing the [Kaggle Dataset](https://www.kaggle.com/djzurawski/us-oil-and-gas-production-june-2008-to-june-2018) with information about Oil and Gas production in the US from June 2008 to June 2018.

```python
# imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Load and explore data

```python
# load data
gas_df = pd.read_csv('data/U.S._natural_gas_production.csv',decimal=',')
gas_df.head(5)
```

```python
print(f"The shape of the gas database is {gas_df.shape}")
```

```python
print(f"Column names: {gas_df.columns}")
```

```python
gas_df.info()
```

```python
#Change the month col to date/time format
gas_df['Month'] = pd.to_datetime(gas_df['Month'])
gas_df.Month.dtype
```

```python
gas_df['Month'].dt.year.head()
```

```python
gas_df['Month'].dt.month.tail()
```

---

## Yearly Gas Production

### How much gas has been produced yearly by each US state and the US as a whole?

```python
gas_df.head(3)
```

```python
# Yearly gas by state and in total
yearly = gas_df.groupby(gas_df.Month.dt.year).sum()
yearly
```

```python
yearly.plot(use_index=True,y='U.S.').set_xlabel('Year')
```

```python
yearly.filter(items=['U.S.']).plot(kind='bar').set_xlabel('Year');
```

```python
filtered_gas_df= yearly[np.logical_and(yearly.index>2008, yearly.index<2018)]
```

```python
filtered_gas_df
```

## Comparing with Crude Oil Production

```python
oil_df = pd.read_csv('data/U.S._crude_oil_production.csv',parse_dates=['Month'])
oil_df.head(2)
```

```python
oil_df['Month'].dtypes
```

---

## Yearly Oil production

```python
yearly_oil = oil_df.groupby(gas_df.Month.dt.year).sum()
yearly_oil
```

```python
yearly_oil.columns=yearly_oil.columns.str.strip()
```

```python
yearly_oil.filter(items=['U.S. Crude Oil']).plot(kind='bar').set_xlabel('Year');
```

```python
# remove incomplete years
yearly_oil_filtered = yearly_oil[np.logical_and(yearly_oil.index>2008,yearly_oil.index<2018)]
yearly_oil_filtered
```

## Merging Oil and Gas Production

```python
total_gas = pd.DataFrame(filtered_gas_df['U.S.'])
total_gas.rename(columns={'U.S.':'Gas'},inplace=True)
total_gas
```

```python
total_oil = pd.DataFrame(yearly_oil_filtered['U.S. Crude Oil'])
total_oil.rename(columns={'U.S. Crude Oil':'Crude Oil'},inplace=True)
total_oil
```

```python
merged = pd.concat([total_oil,total_gas],axis=1)
```

```python
merged
```

```python
plot = merged.plot(kind="bar")
plot.set_xlabel("Year")
plot.legend(['Gas (Millions of Cubic feet)', 'Crude Oil (Thousands of barrels)']);
```
